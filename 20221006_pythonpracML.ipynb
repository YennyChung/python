## Imbalanced Data Sampling
- 분류 (Classification)
- 데이터 비율이 크게 차이 발생한 경우 한쪽에 맞춰서 데이터를 늘리거나 줄이는 작업
  - Under sampling: 데이터의 비율이 적은 값에 맞춰서 데이터를 줄여주는 작업
    - Random Under Sampling: 비율이 높은 쪽의 데이터를 무작위 추출해서 줄임
    - Tomek's Link: 서로 다른 값의 인접한 항목을 묶어서 많은 쪽의 데이터를 삭제
      - 데이터의 비율이 한쪽이 적더라도 분류 경계가 명확해짐 (분류모델성능 증가)
      - 범주형 항목이 다수 있거나, 분류 경계가 매우 모호한 경우
    - CNN (Condensed Nearest Neighbor): 비율이 많은 쪽 데이터에서 밀도가 높은 부분의 데이터 삭제
    - One Sided Selection (Tomek's Link + CNN)
  - Over sampling: 데이터의 비율이 높은 값에 맞춰서 데이터를 생성해주는 작업
    - Random Over Sampling: 비율이 적은 데이터를 무작위로 생성해서 많은 쪽으로 맞춤
    - SMOTE (Synthetic Minority Over Sampling Technique): KNN 알고리즘을 이용해 비율이 적은 쪽의 데이터를 K개의 인접갯수만큼 묶은 뒤 묶인 데이터 내 중심점을 찾아서 데이터 생성
    - ADASYN (Adaptive Synthetic Sampling): SMOTE기법을 이용해서 데이터를 생성하는 단계에서 임의의 값 (Noise, 감마밀도함수)을 더해서 더 사실적인 데이터를 생성하는 기법

- 공정: 불량여부 Y -> 정상 950, 불량 50
  - Under Sampling: 정상 50, 불량 50

```python
# 11. 
from sklearn.preprocessing import MinMaxScaler

def function1(row):
    if row >= df['Y1A'].quantile(0.9):
        return 1
    else:
        return 0
    
df['Target'] = df['Y1A'].apply(function1)
df['Target'].value_counts()

X = df[['A1', 'A2', 'A3', 'A4', 'A5', 'A6']]
Y = df['Target']

over_sampler = SMOTE()
over_sampler.fit(X, Y)
X_over, Y_over = over_sampler.fit_resample(X, Y)
Y_over.value_counts()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

# 숫자가 처리되는 Process 구축 (1. 평균값으로 결측 처리 -> 2. 스케일링 수행)
numeric_pipe = make_pipeline(SimpleImputer(strategy = 'mean'),
                            MinMaxScaler())
numeric_pipe

# 숫자는 숫자끼리, 문자는 문자끼리 처리가 되도록 파이프를 구성
numeric_list = X.describe().columns.tolist()
numeric_list

preprocessing_pipe = make_column_transformer((numeric_pipe, numeric_list))
preprocessing_pipe

# 특성공학학습
model_pipe = make_pipeline(preprocessing_pipe, DecisionTreeRegressor())
model_pipe

grid_model = GridSearchCV(model_pipe, cv=5, 
            param_grid = { }, n_jobs = -1)

grid_model.fit(X_train,Y_train)

best_model = grid_model.best_estimator_

Y_train_pred = best_model.predict(X_train) 
Y_train_pred
Y_test_pred = best_model.predict(X_test) 
Y_test_pred

import pickle
pickle.dump(grid_model, open('model_target.sav', 'wb'))
```
