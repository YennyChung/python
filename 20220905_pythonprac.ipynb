### Python Review
- 변수: 데이터를 담는 공간
- 데이터 타입: 
    - 연속형(int/float64)
    - 범주형(str/object; 범주형)
- 데이터 구조:
    - **List []: 데이터의 순서 존재/중복허용/추가삭제변경 가능***
    - Tuple (): 데이터의 순서 존재/중복허용/추가삭제변경 불가능***
    - Set {}: 데이터의 순서 X/중복X/추가,삭제,변경 가능
    - Dictionary {key:value}: 키-값의 형태/추가,삭제,변경 가능
- 구문:
    - 조건문 (if): 조건에 따라 사용자가 구성한 코드 수행
    - 반복문 (while/for): 특정 횟수/범위에 따라 사용자가 구성한 코드 수행
    - 들여쓰기 (문장의 종속관계 표현 위함)
- 함수(def):
    - 반복사용되는 구문/기능들을 하나의 약어로 정의하여 사용
- 라이브러리: 특정 목적을 수행하기 위해 모아놓은 함수의 집합

### Python 라이브러리
- NumPy(Numeric+Python): 수학, 통계연산과 관련된 함수 집합
- **Pandas(Panel+DataSet): 정형데이터(excel, csv, sql, ..)의 통계량 계산과 전처리 함수의 집합
- Matplot(Matlab + Plot): 연속형 자료들을 시가고하하는 함수들의 집합
- Seaborn: 정형데이터 통계량 시각화 함수들의 집합
- Plotly: 동적 시각화 함수들의 집합
- Missingno: 결측값 시각화 함수들의 집합
- Scipy (Science + Python): 응용통계 (가설검정/시계열분석/회귀분석) 분석함수들의 집합
- Scikit (Science Python Toolkit Learning): 기계학습에 관련한 함수들의 집합 (정형데이터)
- Tensorflow (Tensor + Flow); 비정형데이터(이미지, 영상, 자연어, 신호, etc)를 처리하는 함수들의 집합 + 신경망
    * Scalar (점) - Vector (선) - Matrix (3차원) - Tensor (n차의 차원)
- BS4: 스크래핑 (web)데이터를 수집할 때 사용하는 함수들의 집합
- Selenium: 프로그램 진단; 프로그램 자동화 함수들의 집합
- PyCaret: Auto Machine Learning

- Flask: 웹구현
- Dash: 웹구현 (데이터 기반)

```python
# Bank 항목의 빈도수가 높은 항목들의 순서대로 리스트를 추출
bank_order = data['Bank'].value_counts().index.tolist()
```

```python
bank_order
```

```python
cond1 = data['Bank'].isin(bank_order[0:5])  # 리스트 내 항목에 대해서만 추출하는 함수
cond1
```

```python
data.loc[cond1]  # bank항목에서 빈도수가 높은 상위의 5개 항목에 대해 추출
```

```python
# 요약; Pivot table
# 제품의 종류에 따라서 월렌탈비용의 통계량
data.pivot_table(index = 'Product_Type', values = 'Amount_Month', aggfunc='mean')
# index에는 구분하고 싶은 값 / values에는 계산하고 싶은 값 / aggfunc에는 계산하고자 하는 통계량
data.pivot_table(index = ['Channel','Product_Type'], values = 'Amount_Month', aggfunc = 'sum')
```
