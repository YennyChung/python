```python
import pandas as pd

df_member = pd.read_csv('1_store_member.csv')
df_product = pd.read_csv('2_store_product.csv')
```

```python
# 3)
df_product['공급일자'] = pd.to_datetime(df_product['공급일자'], format = '%Y%m%d')
df_product['공급일자_yy'] = df_product['공급일자'].dt.strftime('%Y')
df_product['공급일자_ym'] = df_product['공급일자'].dt.strftime('%Y-%m')
df_product['공급일자_day'] = df_product['공급일자'].dt.day_name()

df_product
```

```python
# 6)
cond1 = df_product['공급일자_day'].isin(['Saturday', 'Sunday'])
df_weekend = df_product.loc[cond1]
print(df_weekend)
## 6)-1 구매금액의 분포가 정규분포를 따르는지 확인
# 귀무: 구매금액의 분포가 정규분포와 같다.
# 대립: 구매금액의 분포가 정규분포와 같지 않다.

import scipy.stats as stats
stats.normaltest(df_weekend['구매금액'])
# 대립가설이 참
```

```python
# 두 집단 (토요일과 일요일)의 구매금액의 평균의 차이가 있는지?
cond1 = df_weekend['공급일자_day']=='Sunday'
cond2 = df_weekend['공급일자_day']=='Saturday'

stats.ranksums(df_weekend.loc[cond1]['구매금액'], df_weekend.loc[cond2]['구매금액'])
```
