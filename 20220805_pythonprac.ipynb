# lambda() 함수와 filter() 함수의 차이

```python
a = list(range(1,11))
check_list = lambda a: a % 2 == 0

a_even = list()
for i in a:
    if(check_list(i)==True):
        a_even.append(i)

print(a_even)
```

### 함수를 정의해서 대문자로 만들기

```python
def to_upper(i):
    return i.upper()

a = ['a', 'b', 'c', 'd']

index = 0
for x in a:
    a[index] = to_upper(x)
    index += 1

print(a)
```
