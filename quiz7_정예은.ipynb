문제 4

- 데이터 로드

```python
heart_df = pd.read_csv("heart.csv")
heart_df = pd.DataFrame(heart_df)
heart_df
```

- 데이터의 결측치 여부 및 열별 결측치 갯수 파악

```python
heart_df.isnull()
```

```python
heart_df.isnull().sum()
```

- chestpain별 평균 chol 수치

```python
heart_df.groupby(heart_df['ChestPain']).mean()['Chol']
```

---

```python
heart_df.groupby('ChestPain').mean()['Chol']
```

- pivot_table을 이용해서 AHD, sex 행별로 chol 수치를 나타내는 표 만들기

```python
pivot_df = pd.pivot_table(heart_df, index = ['AHD', 'Sex'], values = 'Chol')
pivot_df
```

문제 9

```python
left = pd.DataFrame({'key': ['foo', 'bar', 'poo'], 'lval': [1, 2, 3]})
right = pd.DataFrame({'key': ['bar', 'foo', 'woo'], 'rval': [4, 5, 6]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how = 'left')
```

---

```python
pd.merge(left, right, on = 'key', how = 'left')  # merge 기준을 key (공통으로 들어가는 key)로 설정하기
```

문제 10

```python
pd.merge(left, right, on = 'key', how = 'inner')
```
