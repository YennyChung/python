문제 5

```python
t = np.arange(0,12,0.1)
y = np.sin(t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of Sinewave')
plt.show()
```
